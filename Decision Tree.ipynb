```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = np.array(
[['Asset Flip', 100, 1000],
['Text Based', 500, 3000],
['Visual Novel', 1500, 5000],
['2D Pixel Art', 3500, 8000],
['2D Vector Art', 5000, 6500],
['Strategy', 6000, 7000],
['First Person Shooter', 8000, 15000],
['Simulator', 9500, 20000],
['Racing', 12000, 21000],
['RPG', 14000, 25000],
['Sandbox', 15500, 27000],
['Open-World', 16500, 30000],
['MMOFPS', 25000, 52000],
['MMORPG', 30000, 80000]
])
  
# print the dataset
print(data) 
```

```python
x = data[:,1:2].astype(int)
y = data[:,2].astype(int)
```

```python
x, y
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tree= DecisionTreeClassifier()
```

```python
tree.fit(x,y)
```

```python
tree.predict([[3750]])
```

```python
xgrid= np.arange(np.min(x), np.max(x), .01)
xgrid= xgrid.reshape(len(xgrid),1)
```

```python
plt.scatter(x,y, c= 'r')
plt.plot(xgrid, tree.predict(xgrid))
plt.xlabel('Production cost')
plt.ylabel('Profit')
plt.title('Decision Tree Model')
```

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(50,45))
plot_tree(tree, filled=True, feature_names=['Production Cost'])
plt.show()
```
